```python
import pandas as pd
import numpy as np
from scipy import stats

# Load the dataset
df = pd.read_csv('ibm_stock.csv')

# Calculate Z-scores for 'Close'
z_scores = np.abs(stats.zscore(df['Close']))

# Define a stricter threshold (e.g., Z > 2)
threshold = 2
df_no_outliers = df[z_scores < threshold]

# Save the cleaned data
df_no_outliers.to_csv('ibm_stock_Close_clean_zscore.csv', index=False)

# Summary
print(f"Original rows: {df.shape[0]}")
print(f"Rows after removing outliers using Z-score: {df_no_outliers.shape[0]}")
print("Saved as 'ibm_stock_Close_clean_zscore.csv'")
```
